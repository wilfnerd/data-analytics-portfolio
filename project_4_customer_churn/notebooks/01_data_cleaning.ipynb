```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.shape
df.info()
df['Churn'].value_counts(normalize=True)
```

```python
# Convert TotalCharges to numeric (coerce blanks to NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'].isna().sum()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df['Churn_binary'] = df['Churn'].map({'No': 0, 'Yes': 1})

df['Churn_binary'].value_counts(normalize=True)
```

```python
df.drop(columns=['customerID', 'Churn'], inplace=True)
```

```python
binary_cols = [
    'Partner', 'Dependents', 'PhoneService',
    'PaperlessBilling'
]

for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype('category')
```

```python
categorical_cols = [
    'gender', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaymentMethod'
]

df = pd.get_dummies(df, columns=categorical_cols, drop_first=False)
```

```python
df.shape
df.isna().sum().sort_values(ascending=False).head()
```

```python
df.to_csv('../data/telco_clean.csv', index=False)
```

## Data Cleaning Summary

- Converted TotalCharges to numeric and handled missing values
- Encoded churn as binary target variable
- Dropped non-informative identifiers
- Encoded categorical features using one-hot encoding
- Dataset prepared for exploratory analysis and modeling
